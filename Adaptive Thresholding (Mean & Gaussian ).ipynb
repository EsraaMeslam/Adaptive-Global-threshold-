```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread('sudoku.png',0)


_, th1 = cv.threshold(img, 127, 255, cv.THRESH_BINARY) # make pxs less than threshold (black) else (white) 
th2 = cv.adaptiveThreshold(img, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, 11, 2)#Mean thresoled
th3 = cv.adaptiveThreshold(img, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY, 11, 2)#Gaussian threold

plt.figure(figsize=(20, 10))

plt.subplot(1, 4, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(th1, cmap='gray')
plt.title('THRESH_BINARY')
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(th2, cmap='gray')
plt.title('ADAPTIVE_THRESH_MEAN_C')
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(th3, cmap='gray')
plt.title('ADAPTIVE_THRESH_GAUSSIAN_C')
plt.axis('off')

plt.show()
```

a single global threshold value is used to classify all pixels in an image into foreground (object) or background based on their intensity. However, in some images where lighting conditions vary across the image, a single global threshold might not be effective.

## ADAPTIVE_THRESH_MEAN_C

adaptive thresholding>> calculating different thresholds for different regions of the image, considering local neighborhood regions around each pixel

## ADAPTIVE_THRESH_GAUSSIAN_C

 is particularly effective when the lighting conditions of the image are not uniform across the entire image or when there's a presence of noise, as it adapts locally to these variations.
