```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread('black&white.jpg', -1)  # read image

#parameters(img,threshold val,max px n the image ,type of threshold) ,Return (True OR False),If Trur>>Image after Applying the thresold 
_, th1 = cv.threshold(img, 50, 255, cv.THRESH_BINARY)  # make pxs less than threshold (black) else (white) 
_, th2 = cv.threshold(img, 200, 255, cv.THRESH_BINARY_INV)  # make pxs less threshold 200 (white) else (black)
_, th3 = cv.threshold(img, 127, 255, cv.THRESH_TRUNC) #make pxs less than threshold =threshold else be the same
_, th4 = cv.threshold(img, 127, 255, cv.THRESH_TOZERO)#make pxs more than threshold =0 else be the same
_, th5 = cv.threshold(img, 127, 255, cv.THRESH_TOZERO_INV)#make pxs less than threshold =0 else be the same



plt.figure(figsize=(20, 10))

plt.subplot(2, 3, 1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(2, 3, 2)
plt.imshow(th1, cmap='gray')
plt.title('THRESH_BINARY')
plt.axis('off')

plt.subplot(2, 3, 3)
plt.imshow(th2, cmap='gray')
plt.title('THRESH_BINARY_INV')
plt.axis('off')

plt.subplot(2, 3, 4)
plt.imshow(th3, cmap='gray')
plt.title('THRESH_TRUNC')
plt.axis('off')

plt.subplot(2, 3, 5)
plt.imshow(th4, cmap='gray')
plt.title('THRESH_TOZERO')
plt.axis('off')

plt.subplot(2, 3, 6)
plt.imshow(th5, cmap='gray')
plt.title('THRESH_TOZERO_INV')
plt.axis('off')

plt.tight_layout() 
plt.show()
```
